# Ejemplo funcion


```python
def fib(n):    # write Fibonacci series up to n
     """Print a Fibonacci series up to n."""
     a, b = 0, 1
     while a < n:
         print(a, end=' ')
         a, b = b, a+b
     print()
```

```python
# Para llamar la función que acabamos de definir:
fib(70000)
```

```python
import pandas as pd
# cargamos el dataset
dataset_keys = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
dataset = pd.read_csv('pima-indians-diabetes.csv', names=dataset_keys)
# mostramos el dataset
display(dataset)
```

```python
# problema supervisado: dividimos entre las variables independientes (X) y la dependiente (Outcome -> y)

# dividimos en inputs y outputs
values = dataset.values
X = values[:,0:8]
y = values[:,8]

# estandarizamos los datos
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
# podemos usar una función para dividir entre train y test:

from sklearn.model_selection import train_test_split
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)
```

```python
# vemos que el 80% de los datos son de test y el 20% de train:
print("número de muestras de train: " + str(len(X_scaled_train)))
print("número de muestras de test: " + str(len(X_scaled_test)))
```

```python
# usaremos un modelo support vector machines (svm) 
from sklearn import svm

model = svm.SVC(random_state=42)

# con la funcion fit entrenamos el modelo (datos de train)
model.fit(X_scaled_train, y_train)
```

```python
# una vez hecho el entrenamiento, podemos clasificar los datos nuevos (test)
y_pred = model.predict(X_scaled_test)
print(y_pred)
```

```python
# visualizamos la matriz de confusión 

from sklearn import metrics

cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
```

## Setup

```python
# librerías necesarias para el ejemplo:
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# usamos el mismo dataset que con scikit-learn:
# Utiliza wget para descargarte el notebook
!wget https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv
```

```python
import pandas as pd
# cargamos el dataset
dataset_keys = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
dataframe = pd.read_csv('pima-indians-diabetes.csv', names=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'])
# mostramos el dataframe
display(dataframe)
```

```python
# vamos a dividir los datos: 80% para el entrenamiento y los otros para test
test_dataframe = dataframe.sample(frac=0.2, random_state=1337)
train_dataframe = dataframe.drop(test_dataframe.index)

print(
    "Using %d samples for training and %d for validation"
    % (len(train_dataframe), len(test_dataframe))
)
```

## Preparing the data

Let's download the data and load it into a Pandas dataframe:



Let's split the data into a training and validation set:

Let's generate `tf.data.Dataset` objects for each dataframe:

```python

def dataframe_to_dataset(dataframe):
    dataframe = dataframe.copy()
    labels = dataframe.pop("Outcome")
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe), labels))
    ds = ds.shuffle(buffer_size=len(dataframe))
    return ds


train_ds = dataframe_to_dataset(train_dataframe)
test_ds = dataframe_to_dataset(test_dataframe)
```

Each `Dataset` yields a tuple `(input, target)` where `input` is a dictionary of features
and `target` is the value `0` or `1`:

```python
for x, y in train_ds.take(1):
    print("Input:", x)
    print("Target:", y)
```

Let's batch the datasets:

```python
train_ds = train_ds.batch(32)
test_ds = test_ds.batch(32)
```

```python
train_ds
```

```python
from tensorflow.keras.layers import IntegerLookup
from tensorflow.keras.layers import Normalization
from tensorflow.keras.layers import StringLookup


def encode_numerical_feature(feature, name, dataset):
    # Create a Normalization layer for our feature
    normalizer = Normalization()

    # Prepare a Dataset that only yields our feature
    feature_ds = dataset.map(lambda x, y: x[name])
    feature_ds = feature_ds.map(lambda x: tf.expand_dims(x, -1))

    # Learn the statistics of the data
    normalizer.adapt(feature_ds)

    # Normalize the input feature
    encoded_feature = normalizer(feature)
    return encoded_feature


```

## Build a model

With this done, we can create our end-to-end model:

```python
# Numerical features
Pregnancies = keras.Input(shape=(1,), name="Pregnancies")
Glucose = keras.Input(shape=(1,), name="Glucose")
BloodPressure = keras.Input(shape=(1,), name="BloodPressure")
SkinThickness = keras.Input(shape=(1,), name="SkinThickness")
Insulin = keras.Input(shape=(1,), name="Insulin")
BMI = keras.Input(shape=(1,), name="BMI")
DiabetesPedigreeFunction = keras.Input(shape=(1,), name="DiabetesPedigreeFunction")
Age = keras.Input(shape=(1,), name="Age")

all_inputs = [
    Pregnancies,
    Glucose,
    BloodPressure,
    SkinThickness,
    Insulin,
    BMI,
    DiabetesPedigreeFunction,
    Age
]

# Encode Numerical features
Pregnancies_encoded = encode_numerical_feature(Pregnancies, "Pregnancies", train_ds)
Glucose_encoded = encode_numerical_feature(Glucose, "Glucose", train_ds)
BloodPressure_encoded = encode_numerical_feature(BloodPressure, "BloodPressure", train_ds)
SkinThickness_encoded = encode_numerical_feature(SkinThickness, "SkinThickness", train_ds)
Insulin_encoded = encode_numerical_feature(Insulin, "Insulin", train_ds)
BMI_encoded = encode_numerical_feature(BMI, "BMI", train_ds)
DiabetesPedigreeFunction_encoded = encode_numerical_feature(DiabetesPedigreeFunction, "DiabetesPedigreeFunction", train_ds)
Age_encoded = encode_numerical_feature(Age, "Age", train_ds)

all_features = layers.concatenate(
    [
        Pregnancies_encoded,
        Glucose_encoded,
        BloodPressure_encoded,
        SkinThickness_encoded,
        Insulin_encoded,
        BMI_encoded,
        DiabetesPedigreeFunction_encoded,
        Age_encoded,
    ]
)
```

```python

# definimos la red: entrada all_features; salida 1 sola variable (tiene o no tiene diabetes)
x = layers.Dense(32, activation="relu")(all_features)
x = layers.Dropout(0.5)(x)
output = layers.Dense(1, activation="sigmoid")(x)
model = keras.Model(all_inputs, output)

# debemos compilar el modelo, indicando una loss indicada (problema de clasificación binario)
model.compile("adam", "binary_crossentropy", metrics=["accuracy"])
```

Let's visualize our connectivity graph:

```python
#  podemos visualizar el modelo
keras.utils.plot_model(model, show_shapes=True, rankdir="LR")
```

## Train the model

```python
# una vez creado el modelo y compilado ya podemos entrenar la red:
model.fit(train_ds, epochs=50, validation_data=train_ds)
```

We quickly get to 75% validation accuracy.

## Inference on new data

To get a prediction for a new sample, you can simply call `model.predict()`. There are just two things you need to do:

1. wrap scalars into a list so as to have a batch dimension (models only process batches
of data, not single samples)
2. Call `convert_to_tensor` on each feature

```python
# podemos crear nuevos pacientes y evaluar la probabilidad de tener diabetes con el modelo que hemos entrenado:
sample = {
    "Pregnancies": 6,
    "Glucose": 140,
    "BloodPressure": 70,
    "SkinThickness": 35,
    "Insulin": 0,
    "BMI": 31.1,
    "DiabetesPedigreeFunction": 0.64,
    "Age": 50,
}


input_dict = {name: tf.convert_to_tensor([value]) for name, value in sample.items()}
predictions = model.predict(input_dict)

print(
    "This particular patient had a %.1f percent probability "
    "of having diabetes, as evaluated by our model." % (100 * predictions[0][0],)
)
```
